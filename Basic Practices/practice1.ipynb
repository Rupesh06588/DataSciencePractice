# Create dataset of 5 student with name, makes, age and grades. Print average marks and filter who scored more than 70

```python
import pandas as pd
```

```python
data = {
    "Name": ["Ram", "Sita", "Andrew", "Alice", "Bob"],
    "Age" : [20, 22, 23, 24, 27],
    "Marks": [80, 78, 65, 92, 50],
    "Grades": ["B", "C", "D", "A", "E"]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
print("Average marks =", df['Marks'].mean())
```

```python
print(df[df["Marks"]>70])
```
