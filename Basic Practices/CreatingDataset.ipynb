# Basic dataset

```python
import pandas as pd

data = {
    "Name" : ["Alice", "Bob", "Charlie", "David"],
    "Age" : [20, 30, 40, 50],
    "Salary": [50000, 60000, 70000, 75000]
    }

df = pd.DataFrame(data)
print(df) #dispays the dataset

print(df.describe()) #basic statistics
```
