```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Customer': ['Alice','Bob','Charlie','David','Eve','Frank','Gita','Hari','Isha','Jay',
                 'Kiran','Lina','Manoj','Nita','Om','Pooja','Qasim','Rita','Sam','Tina'],
    'Age': [25,32,35,40,30,29,22,45,38,27,33,41,36,28,50,24,31,42,39,26],
    'City': ['New York','New York','New York','Los Angeles','Los Angeles','Los Angeles',
             'New York','Los Angeles','New York','Los Angeles',
             'New York','Los Angeles','New York','Los Angeles','New York','Los Angeles',
             'New York','Los Angeles','New York','Los Angeles'],
    'Purchase': [250,300,350,400,500,600,200,700,450,280,330,650,370,260,800,240,310,620,390,270],
    'Gender': ['F','M','M','M','F','F','F','M','F','M','M','F','M','F','M','F','M','F','M','F']
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
print("Data Statistics")
df.describe()
```

```python
# Correlation Matrix
print("Correlation matrix")
numeric_df = df.select_dtypes(include=['Int64', 'float64'])
print(numeric_df.corr())
```

```python
sns.heatmap(numeric_df.corr(), annot= True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(df["Age"], bins=8, kde=True)
plt.title("Age Distribution")
plt.show() 
```

```python
sns.barplot(x= "City", y= "Purchase", data=df)
plt.title("Average Distribution")
plt.show()
```

```python
sns.scatterplot(x='Age', y= 'Purchase', data=df, hue='Gender', s=100)
plt.title("Age vs Purchase by Gender")
plt.show()
```
