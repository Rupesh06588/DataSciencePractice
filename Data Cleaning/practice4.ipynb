# Clean the given data

```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Customer': ['Alice', 'Bob', 'Charlie', 'Alice', 'Eve', 'Frank', None],
    'Age': [25, None, 35, 40, 'Thirty', 29, 33],
    'City': ['NYC', 'new york', 'New York', 'LA', 'Los Angeles', 'la', 'NYC'],
    'Purchase': [250, 300, None, 400, 500, 600, 250],
    'Gender': ['F', 'M', 'Male', 'M', 'F', 'Female', None]
}
```

```python
df = pd.DataFrame(data)
print("Uncleaned Data")
print(df)
```

```python
print(df.isnull())
```

```python
df['Purchase'] = df['Purchase'].fillna(df['Purchase'].mean())
```

```python
df["Age"] = df["Age"].replace({"Thirty": 30})
df["Age"] = pd.to_numeric(df["Age"])
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df = df.dropna()
```

```python
df['City'] = df['City'].str.lower().replace({
    "nyc": "New York",
    "la" : "Los Angeles",
    "los angeles": "Los Angeles",
    "new york": "New York"
})
```

```python
df['Gender'] = df['Gender'].replace({
    "Male": "M",
    "Female": "F",
})
```

```python
df = df.drop_duplicates()
```

```python
print("Cleaned Data")
print(df)
```
