# Given a dataset with missing marks of students, fill missing values with the average marks.

```python
import pandas as pd
students = {
    'Name': ['Ravi', 'Sita', 'Aman', 'Priya'],
    'Marks': [85, None, 92, 78]
}
```

```python
df = pd.DataFrame(students)
```

```python
print(df)
```

```python
print(df["Marks"].isnull())
```

```python
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
```

```python
print(df["Marks"])
```
