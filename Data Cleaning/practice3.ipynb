# Remove duplicate rows from the dataset below

```python
import pandas as pd
```

```python
data = {
    'Product': ['A', 'B', 'C', 'A', 'B'],
    'Price': [100, 200, 300, 100, 200]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
df= df.drop_duplicates()
```
