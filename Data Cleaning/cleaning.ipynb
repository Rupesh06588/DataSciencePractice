# Cleaning data

```python
import pandas as pd
```

```python
data = {
    "Name": ['alice', 'bob', 'mark', 'herry', 'eve'],
    "Age": [25, None, 35, 40, 30],
    "Gender": ['F', 'M', 'Male', 'M', 'F'],
    "Salary": [50000, 60000, None, 80000, 60000]
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Original data:")
print(df)
```

```python
# handle missing values
df["Age"]=df["Age"].fillna(df["Age"].mean(), inplace = True)
df["Salary"]= df['Salary'].fillna(df["Salary"].mean(), inplace= True)
```

```python
# Standardize the gender column
df["Gender"] = df["Gender"].replace({"Male": 'M', "Female": "F"})
```

```python
print("Cleaned Data:")
print(df)
```
