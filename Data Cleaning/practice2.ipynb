# Standardize the column city where the values are NYC, New York, new york --> All should become New York

```python
import pandas as pd
cities = {
    "city" : ['NYC', 'New York', 'new york', 'nyc']
}
```

```python
df = pd.DataFrame(cities)
print(df)
```

```python
df["city"] = df["city"].str.lower().replace({"nyc": "New York", "new york": "New York"})
```
