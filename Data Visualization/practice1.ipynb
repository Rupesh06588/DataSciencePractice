```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Customer': ['Alice','Bob','Charlie','David','Eve','Frank','Gita','Hari','Isha','Jay',
                 'Kiran','Lina','Manoj','Nita','Om','Pooja','Qasim','Rita','Sam','Tina'],
    'Age': [25,32,35,40,30,29,22,45,38,27,33,41,36,28,50,24,31,42,39,26],
    'City': ['New York','New York','New York','Los Angeles','Los Angeles','Los Angeles',
             'New York','Los Angeles','New York','Los Angeles',
             'New York','Los Angeles','New York','Los Angeles','New York','Los Angeles',
             'New York','Los Angeles','New York','Los Angeles'],
    'Purchase': [250,300,350,400,500,600,200,700,450,280,330,650,370,260,800,240,310,620,390,270],
    'Gender': ['F','M','M','M','F','F','F','M','F','M','M','F','M','F','M','F','M','F','M','F']
}

df = pd.DataFrame(data)
print(df)
```

```python
sns.histplot(df['Purchase'], kde=True, bins=5)
plt.title("Distribution of purchase")
plt.show()
```

```python
sns.barplot(x= 'Gender', y= 'Purchase', data = df)
plt.title("Average Purchase by Gender")
plt.show()
```

```python
sns.barplot(x= 'City', y= 'Purchase', data = df)
plt.title("Average Purchase by City")
plt.show()
```

```python
sns.scatterplot(x='Age', y='Purchase', hue='Gender', data= df, s= 50)
plt.title("Age vs Purchase by Gender")
plt.show()
```

```python
#linear model plot, this fits the regression line to the scatter plot
sns.lmplot(x= 'Age', y= 'Purchase', hue='Gender', data= df, height= 6, aspect= 1.2)
plt.title('Trend line: Age vs Purchase by Gender')
plt.show()
```
