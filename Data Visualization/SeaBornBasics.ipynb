```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# sample dataset
data = {
    'city': ['newyork', 'newyork', 'los angeles', 'los angeles', 'los angeles'],
    'purchase': [250, 300, 400, 500, 600],
    'Gender': ['F', 'M', 'M', 'F', 'F']
}
df = pd.DataFrame(data)
print(df)
```

```python
sns.barplot(x= 'city', y='purchase', data=df)
plt.title("Average Purchase by city")
plt.show()
```

```python
sns.histplot(df['purchase'], bins=5, kde=True)
plt.title("Purchase Distribution")
plt.show()
```
