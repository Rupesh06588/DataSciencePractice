```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cleaned_titanic_dataset")
df.head()
```

```python
y = df["survived"]
x = df.drop(columns = ["survived"])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

# Metrics Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report
# print("Accuracy_score: ", accuracy_score(y_test, y_pred))
print("classification_report: \n ", classification_report(y_test, y_pred))
```

```python

import matplotlib.pyplot as plt
# Take one feature (Age)
x_age = x_train[["age"]]

model_age = LogisticRegression()
model_age.fit(x_age, y_train)

# Create smooth curve
X_test_vals = np.linspace(x_age.min(), x_age.max(), 300).reshape(-1,1)
y_prob = model_age.predict_proba(X_test_vals)[:,1]

plt.plot(X_test_vals, y_prob, color="blue")
plt.title("Logistic Regression (Sigmoid Curve)")
plt.xlabel("Age")
plt.ylabel("Survival Probability")
plt.show()
```
