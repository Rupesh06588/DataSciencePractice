```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
missing_data = df.isnull()
for columns in df:
    print(missing_data[columns].value_counts())
    print(' ')
```

# Fixing Null Values

```python
df["age"] = df["age"].fillna(df["age"].mean())
df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])
df["embark_town"] = df["embark_town"].fillna(df["embark_town"].mode()[0])

df.drop("deck", axis=1, inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.drop(["embark_town", "alive", "class", "who", "adult_male"], axis=1, inplace=True)
# this is because these are the data that are already in another format.
```

# one hot encoding

```python
df["sex"] = df["sex"].map({"male": "0", "female": "1"})
df = pd.get_dummies(df, columns=["embarked"], drop_first=True )
df.head()
```

# Export the cleaned Dataset

```python
df.to_csv("cleaned_titanic_dataset", index=False)
```
